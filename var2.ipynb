```python
import pandas as pd
df = pd.read_csv (r'E:\Irene\ghrelin2.csv')
```

```python
df.describe()
```

```python
df.info()
```

```python
df.head(8)
```

```python
df = df.drop(['Cod'], axis=1)
df = df.drop(['detalii'], axis=1)
df = df.drop(['Ghr muc /5HPF'], axis=1)
df = df.drop(['Ghr muc /HPF'], axis=1)
df = df.drop(['ghrelin rank'], axis=1)
df = df.rename(columns={'Ghr muc/10HPF': 'Ghr'})
df = df.rename(columns={'Inflam rank': 'Inflam'})
df = df.rename(columns={'Infiltr mono': 'mono'})
df = df.rename(columns={'Infiltr neutrofile': 'neutro'})
df = df.rename(columns={'Grup progn': 'progn'})
df = df.rename(columns={'Hp rank': 'Hp_rk'})
df = df.rename(columns={'Loc mucoasa': 'mucoasa'})
```

```python
df['Sex'].replace(['F','M'],[0,1],inplace=True)
```

```python
df['Micro'].replace(['fusiform','mixt','epitelioid'],[1,2,3],inplace=True)
```

```python
df['pleomorfism'].replace(['redus','mod','ridicat'],[1,2,3],inplace=True)
```

```python
df['AM'].replace(['low','high'],[0,1],inplace=True)
```

```python
df['CD117'].replace(['neg','p'],[0,1],inplace=True)
df['DOG1'].replace(['neg','p'],[0,1],inplace=True)
df['CD34'].replace(['neg','p'],[0,1],inplace=True)
```

```python
df['progn'].replace(['1','2','3a','3b','4','5','6a','6b'],[1,2,3,3.5,4,5,6,6.5],inplace=True)
```

```python
df['Distributie'].replace(['1/3 baz','2/3 sup','random'],[1,2,3],inplace=True)
```

```python
df['mucoasa'].replace(['corp','antrum'],[0,1],inplace=True)
```

```python
from scipy.stats import chi2_contingency
contigency= pd.crosstab(df['T'], df['progn'])
contigency_pct = pd.crosstab(df['T'], df['progn'], normalize='index')
contigency_pct
c, p, dof, expected = chi2_contingency(contigency) # Print the p-value
print(p)
plt.figure(figsize=(12,8)) 
sns.heatmap(contigency, annot=True)
```
