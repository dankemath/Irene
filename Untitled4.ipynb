```python
from scipy import stats
def correlation(corr, dataset, threshold1, threshold2):
    for i in range(len(corr.columns)):
        for j in range(i):
            if ((corr.iloc[i, j] > threshold1) and (corr.iloc[i, j] <= threshold2)) or ((corr.iloc[i,j]<-threshold1) and (corr.iloc[i, j] >= -threshold2)):
                print(corr.columns[i],'is correlated with',corr.columns[j])                                               
                pearson_coef, p_value = stats.pearsonr(df[corr.columns[i]], df[corr.columns[j]])
                print("The Correlation Coefficient is", pearson_coef, " with a P-value of P =", p_value)
print("PEARSON")               
print("STRONG LINEAR CORRELATION")             
correlation(corrMatrix1,df,0.7,1)
print("LINEAR CORRELATION")  
correlation(corrMatrix1,df,0.5,0.7)
print("WEAK LINEAR CORRELATION")  
correlation(corrMatrix1,df,0.3,0.5)
```
