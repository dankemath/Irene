```python
import pandas as pd
pd.set_option('display.max_columns', None)
df = pd.read_csv (r'E:\Irene\BD.csv')
```

```python
df.info()
```

```python
df=df.rename(columns={"Nr. Caz": "nr_caz", "Registru Histopat": "Histopat", "Micro subtip":"Micro", "Ki-67 cutoff": "Ki67","Grup progn":"progn"})       
```

```python
df.info()
```

```python
df["Sex"] = df["Sex"].astype('category')
df["Localizare"] = df["Localizare"].astype('category')
df["Micro"]=df["Micro"].astype('category')
df["Pleomorfism"]=df["Pleomorfism"].astype('category')
df["AM"]=df["AM"].astype('category')
df["CD117"]=df["CD117"].astype('category')
df["DOG1"]=df["DOG1"].astype('category')
df["CD34"]=df["CD34"].astype('category')
df["Ki67"]=df["Ki67"].astype('category')
df["progn"]=df["progn"].astype('category')
df["Mucoasa"]=df["Mucoasa"].astype('category')
df["Seroasa"]=df["Seroasa"].astype('category')
df["Hemoragie"]=df["Hemoragie"].astype('category')
df["Necro"]=df["Necro"].astype('category')
df["Degen"]=df["Degen"].astype('category')
```

```python
df.info()
```

```python
df["Sex"] = df["Sex"].cat.codes
df["Localizare"]=df["Localizare"].cat.codes
df["Micro"]=df["Micro"].cat.codes
df["Pleomorfism"]=df["Pleomorfism"].cat.codes
df["AM"]=df["AM"].cat.codes
df["CD117"]=df["CD117"].cat.codes
df["DOG1"]=df["DOG1"].cat.codes
df["CD34"]=df["CD34"].cat.codes
df["Ki67"]=df["Ki67"].cat.codes
df["progn"]=df["progn"].cat.codes
df["Mucoasa"]=df["Mucoasa"].cat.codes
df["Seroasa"]=df["Seroasa"].cat.codes
df["Hemoragie"]=df["Hemoragie"].cat.codes
df["Necro"]=df["Necro"].cat.codes
df["Degen"]=df["Degen"].cat.codes
df=df.drop("nr_caz",axis=1)
```

```python
# Seaborn visualization library
import seaborn as sns# Create the default pairplot
import matplotlib.pyplot as plt
imag=sns.pairplot(df)
imag.savefig("E:\Irene\output1.png")
```

```python
corrMatrix1 = df.corr(method='pearson')
fig, ax = plt.subplots(figsize=(20,20))    
a=sns.heatmap(corrMatrix1, annot = True, ax=ax, center=0)
a
corrMatrix1.to_csv(r'E:\Irene\new\correlation_pearson1.csv')
a.figure.savefig(r'E:\Irene\\new\heatmap_pearson1.png',figsize=(20,20))
```
